### Importing Libraries

```python
from sklearn.datasets import load_iris
import  pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing iris dataset 

```python
iris = load_iris()
df= pd.DataFrame(data=iris.data,columns=iris.feature_names)
df['Species'] = iris.target
```

```python
df.head(10)
```

### Task B

```python
print("Number of attributes in dataset: ",df.shape[1])
```

```python
print("Number of samples: ",df.shape[0])
```

```python
#What are the value range for each aLribute?

print("*==Maximum value range==*")
for col in df.columns:
    
    print(f"{col} maximum value",df[col].max())
print("\n")
print("*==Minimum value range==*")
for col in df.columns:
    
    print(f"{col} maximum value",df[col].min())
print("\n")
print("*==Median value range==*")
for col in df.columns:
    
    print(f"{col} maximum value",df[col].median())

plt.figure(figsize=(12, 8))
plt.boxplot(df.values, labels=df.columns)
plt.title('Basic Boxplot of Iris Dataset')
plt.ylabel('Values')
plt.show()
```

```python
#What are the class names?

print("classes names:\n",df.columns.tolist())
```

```python
#How many samples per class?
for col in df.columns:
    print(f"samples per class({col}):",df[col].value_counts())
print("Total samples: \n",df.count())
```

```python
# Find out which two classes have more overlapping tendency, and which
# one is likely to be a separate class as can been seen in the ploLing.

sns.pairplot(df,hue='Species')
plt.show()
```

* Species 0 with sepal length and sepal width shows more overlapping tendency.
* Petal length & Petal width shows best seperation
