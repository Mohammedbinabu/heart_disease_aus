## Data Exploration

#### Heart Disease Dataset Information

#### Features

- **Age** : Integer (in years)  
- **Sex** :  
  - 1 = Male  
  - 0 = Female  

- **Chest Pain Type** :  
  - 1 = Typical angina  
  - 2 = Atypical angina  
  - 3 = Non-anginal pain  
  - 4 = Asymptomatic  

- **Resting Blood Pressure** : (in mm Hg on admission to the hospital)  

- **Serum Cholesterol** : (in mg/dl)  

- **Fasting Blood Sugar (>120 mg/dl)** :  
  - 1 = True  
  - 0 = False  

- **Resting Electrocardiographic Results** :  
  - 0 = Normal  
  - 1 = ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)  
  - 2 = Probable or definite left ventricular hypertrophy by Estes' criteria  

- **Thalach** : Maximum heart rate achieved  

- **Exercise Induced Angina** :  
  - 1 = Yes  
  - 0 = No  

- **Oldpeak** : ST depression induced by exercise relative to rest  

- **Slope (peak exercise ST segment)** :  
  - 1 = Upsloping  
  - 2 = Flat  
  - 3 = Downsloping  

- **Number of Vessels Colored by Fluoroscopy** : Integer (0–3)  

- **Thalassemia (Thal)** :  
  - 3 = Normal  
  - 6 = Fixed defect  
  - 7 = Reversible defect  

---

### Target (Output)

- **0** = No Heart Disease  
- **1** = Heart Disease

### Count and Ratio

```python
#b

# Getting the attributes or classes which are categorical
class_columns = df.select_dtypes(include='object')
class_columns
```

```python
# Observation of class attrivutes.
# Total count for each feature and ratio of each classes within the feature.
for col in class_columns.columns:
    print(f"\n*=== Column: {col} ===*\n")
    summary = (class_columns[col].value_counts().to_frame("Count"))
    summary["Ratio (%)"] = (class_columns[col].value_counts(normalize=True) * 10).round(2)
    print(summary)
    print("-" * 60)

    
    plt.figure(figsize=(5,3))
    summary["Count"].plot(kind="pie", autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors,legend=False)
    plt.title(f"Distribution of {col} (Total: {len(class_columns[col])})")
    plt.ylabel("")  # removes 'count' from y-axis
    plt.show()
```

### Histogram

```python
# histogram for age column:


# Divided the age column into five subcategories, each approximately 10 years
df.age.groupby(pd.cut(df.age, 5)).count().plot(kind="bar")
# Line plot as kernel density estimation of the histogram
df.age.groupby(pd.cut(df.age, 5)).count().plot(kind="line",color="red",marker="*")
plt.tick_params(labelrotation=0)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency/Count")
plt.show()
```

* The biggest group, with over 350 people, falls into the 57-67 age bracket.

* The next largest group, with 300 to 350 people, is in the 48-57 age group.

* he third largest, with 200 to 250 people, is the 39-48 age group.

* Interestingly, the youngest and oldest age categories are the smallest groups in  the dataset.
